```python
import cv2 
import matplotlib.pyplot as plt
import os
```

```python
def load_img(path,type='gray',show=False):
    img = cv2.imread(path)
    
    if type=='gray':
        img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    if type=='rgb':
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    if show == True:
        plt.imshow(img)
        plt.xticks([])
        plt.yticks([])
        plt.show()
    
    return img
    
    
    
```

```python
def detect_face(image,scaleFactor=1.1,minNeighbor=5,color=(0,255,0),width=3,show=True):
    haar = cv2.CascadeClassifier("data/lbpcascade_frontalface.xml")
    faces = haar.detectMultiScale(image,scaleFactor,minNeighbor)
    for (x,y,w,h) in faces:
        image = cv2.rectangle(image,(x,y),(x+w,y+h),color,width)
    
    if show == True:
        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()
    
    return image
```

```python
img  = load_img("data/test5.jpg","rgb",True)
```

```python
image = detect_face(img,color=(0,255,0))
```
