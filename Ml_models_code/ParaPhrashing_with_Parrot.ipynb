```python
from parrot import Parrot 
import torch 
import warnings 
warnings.filterwarnings("ignore")
```

```python
torch.manual_seed(1234)
```

```python
model =Parrot(model_tag="prithivida/parrot_paraphraser_on_T5",use_gpu=False)
```

```python
phrases  = ["You want to be On Trip ?","Where is college ?"]
```

```python
for ph in phrases:
  print("-"*100)
  print(ph)
  print("-"*100)

  phrases_ =  model.augment(input_phrase= ph)

  for ph_ in phrases_:
    print(ph_)
```
